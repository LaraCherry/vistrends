```python
import csv
import spacy

nlp = spacy.load('en')
articles = []
with open('articles.csv', newline='') as csvfile:
    reader = csv.DictReader(csvfile)
    for row in reader:
        articles.append({'headline': row['headline'], 'text': row['article_text']})
```

```python
print(articles[8:9])
```

```python
n = 1
for item in articles:
    np_list_header = []
    np_list_article = []
    ent_list_header = []
    ent_list_article = []
    doc1 = nlp(item['headline'])
    doc2 = nlp(item['text'])
    print('*****Item number %d*****' % n)
    for np in doc1.noun_chunks:
        np_list_header.append(np.text)
    for ent in doc1.ents:
        ent_list_header.append(ent.text)
    for np in doc2.noun_chunks:
        np_list_article.append(np.text)
    for ent in doc2.ents:
        ent_list_article.append(ent.text)
    print('Noun chunks in header: %s' % ', '.join(np_list_header))
    print('NER in header: %s' % ', '.join(ent_list_header))
    print('Noun chunks in article: %s' % ', '.join(np_list_article))
    print('NER in article: %s' % ', '.join(ent_list_article))
    print('\n')
    n+=1
```
